```python
operations = ['subtract', 'add', 'multiply', 'divide']
```

```python
def add_three_node(parent, operation):
    new_node = ThreeNode(operation=operation)
    parent.add_edge(new_node)
    return new_node
```

```python
def add_operations_recursively(node, operations, reapeat):
    if reapeat == 4:
        return
    for operation in operations:
        new_node = add_three_node(node, operation)
        add_operations_recursively(new_node, operations, reapeat + 1)
    
```

```python
def print_tree_recursive(roots):
    for root in roots:
        print(f"Root operation: {root.operation}")
        print_children(root.edges, 1)

def print_children(edges, level):
    for edge in edges:
        print(f"{'  ' * level}Child operation: {edge.operation}")
        print_children(edge.edges, level + 1)
```

```python
roots = [ThreeNode(operation=operation) for operation in operations]

```

```python
for root in roots:
    add_operations_recursively(root, operations, 0)
```

```python
print_tree_recursive(roots)
```
